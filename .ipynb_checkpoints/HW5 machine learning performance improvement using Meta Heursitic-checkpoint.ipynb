# step1 : 載入鐵達尼號的資料並觀察欄位資訊

```python
import pandas as pd
import numpy as np
import random
from sklearn import preprocessing
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
import sys
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('train.csv')
data_ = pd.read_csv('test.csv')
print(data.shape)
print(data.columns)
print(data.info())
```

### 觀察資料間的關係

```python
import seaborn as sns
sns.set(style="ticks", color_codes=True)
sns.pairplot(data)
```

# step2: 前處理 + 選擇我們要的特徵 + 準備我們的目標

```python
# 選擇用於訓練的特徵
features = ['Pclass', 'Sex', 'Age', 'Fare']
x_train = data[features]
x_test = data_[features]
y_train=data['Survived']
```

```python
#檢查缺值
x_train.info()
print('-'*100)
x_test.info()
```

```python
# 補age,fare的空值資料

# 使用平均年齡來填充年齡中的nan值
x_train['Age'] = x_train['Age'].fillna(x_train['Age'].mean())
x_test['Age'] = x_test['Age'].fillna(x_test['Age'].mean())
# 使用票價的均值填充票價中的nan值
x_train['Fare'] = x_train['Fare'].fillna(x_train['Fare'].mean())
x_test['Fare'] = x_test['Fare'].fillna(x_test['Fare'].mean())
print(x_train.shape, x_test.shape)
```

```python
#Normalize + one hot
x_age = x_train['Age'].values.reshape(-1, 1)
x_fare = x_train['Fare'].values.reshape(-1,1)

scaler = preprocessing.MinMaxScaler()

x_train['Age'] = scaler.fit_transform(x_age)
x_train['Fare'] = scaler.fit_transform(x_fare)

x_train = pd.get_dummies(x_train, columns=['Sex', 'Pclass'])

# TEST
x_age = x_test['Age'].values.reshape(-1, 1)
x_fare = x_test['Fare'].values.reshape(-1,1)

x_test['Age'] = scaler.fit_transform(x_age)
x_test['Fare'] = scaler.fit_transform(x_fare)
x_test = pd.get_dummies(x_test, columns=['Sex', 'Pclass'])

print(x_train.shape, x_test.shape)
```

```python
x_train_ = x_train.values.reshape(-1, 7)
y_train_ = y_train.values.reshape(-1, 1)

x_test_ = x_test.values.reshape(-1, 7)
print(x_test_)
print(x_train_)
```

# step3: 建立SVC模型
    選擇以下三種參數做為調整
    [C] : float, optional (default=1.0). Penalty parameter C of the error term.
    [degree] : int, optional (default=3). Degree of the polynomial kernel function (‘poly’). Ignored by all other kernels.
    [gamma] : float, optional (default=’auto’). Kernel coefficient for ‘rbf’, ‘poly’ and ‘sigmoid’. If gamma is ‘auto’ then 
            1/n_features will be used instead.

```python
def model(x, y, c, Gamma, Degree):
    model = SVC(C = c, gamma = Gamma, degree = Degree)
    model.fit(x, y)
    s = cross_val_score(model, x, y, cv=10)
    return model
```

# step4: 建立HC優化器

```python
class HC():
    def __init__(self, c_size, gamma_size,degree_size):
        self.c_size = c_size
        self.gamma_size = gamma_size
        self.degree_size = degree_size
        solution=[]
        # for C
        solution.append((random.uniform(1.0, self.c_size)))
        # for gamma
        solution.append((random.uniform(1.0, self.gamma_size)))
        # for degree
        solution.append((random.randint(1, self.degree_size)))

        self.solution = solution
        self.fitness = 0
    def transition(self):
        counter = random.randint(0,2)
        self.solution_ = self.solution[:]
        
        if counter == 0:
            self.solution_[counter] = random.uniform(1, self.c_size)
        elif counter == 1:
            self.solution_[counter] = random.uniform(1, self.gamma_size)
        else:
            self.solution_[counter] = random.randint(1, self.degree_size)
    def determination(self, fitness_):
        if self.fitness < fitness_:
            self.solution = self.solution_[:]
            self.fitness = fitness_
            print(self.fitness)
        
```

# step5: 訓練模型
    <img src="./svm_per.jpg" width = "500" height = "200" align=center />

```python
hc = HC(10000.0,10000.0, 11)
for i in range(10):
    hc.transition()
    SVMmodel =model(x_train_,y_train_,hc.solution_[0],hc.solution_[1], hc.solution_[2])
    pred_y = SVMmodel.predict(x_train_)
    fitness = accuracy_score(y_train, pred_y)
    hc.determination(fitness)
```

# step6 : 預測測試資料的生存率

```python
SVMmodel =model(x_train_,y_train_,hc.solution_[0],hc.solution_[1])
pred_test_y = SVMmodel.predict(x_test_)
print(pred_test_y)
```
